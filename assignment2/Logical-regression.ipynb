```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
#读入数据集
d4 = pd.read_csv("d4.data", header=None)
d4.head()
```

```python
#将第二列数据中的M转化为0，B转化为1
d4.iloc[:,1].replace(["M","B"], [0,1], inplace=True)
d4.head()
```

```python
"""
第一列数据为人员编号,对分类器无意义，丢弃不用；
第二列数据为诊断结果，作为类别标签；
其他数据作为输入;
将数据转化为array类型。
"""
X = d4.iloc[:,2:].values
Y = d4.iloc[:,1].values
```

```python
#划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#定义逻辑回归模型
def weightInitialization(n_features):
    w = np.zeros((1,n_features))
    b = 0
    return w,b

def sigmoid_activation(result):
    final_result = 1/(1+np.exp(-result))
    return final_result

def model_optimize(w, b, X, Y):
    m = X.shape[0]
    #损失计算
    final_result = sigmoid_activation(np.dot(w,X.T)+b)
    Y_T = Y.T
    cost = (-1/m)*(np.sum((Y_T*np.log(final_result)) + ((1-Y_T)*(np.log(1-final_result)))))
    #梯度计算
    dw = (1/m)*(np.dot(X.T, (final_result-Y.T).T))
    db = (1/m)*(np.sum(final_result-Y.T))
    grads = {"dw": dw, "db": db}
    return grads, cost

def model_predict(w, b, X, Y, learning_rate, no_iterations):
    costs = []
    for i in range(no_iterations):
        grads, cost = model_optimize(w,b,X,Y)
        dw = grads["dw"]
        db = grads["db"]
        #参数更新
        w = w - (learning_rate * (dw.T))
        b = b - (learning_rate * db)
        if (i % 100 == 0):
            costs.append(cost)
    #最终的参数
    coeff = {"w": w, "b": b}
    gradient = {"dw": dw, "db": db}
    return coeff, gradient, costs

def predict(final_pred, m):
    y_pred = np.zeros((1,m))
    for i in range(final_pred.shape[1]):
        if final_pred[0][i] > 0.5:
            y_pred[0][i] = 1
    return y_pred
```

```python
#用训练集训练模型，得到参数
n_features = X_train.shape[1]
w, b = weightInitialization(n_features)
coeff, gradient, costs = model_predict(w, b, X_train, Y_train, learning_rate=0.0001,no_iterations=4500)
```

```python
#用测试集进行测试
w = coeff["w"]
b = coeff["b"]
final_test_pred = sigmoid_activation(np.dot(w,X_test.T)+b)
m_test =  X_test.shape[0]
y_test_pred = predict(final_test_pred, m_test)
```

```python
#得到混淆矩阵
cm = confusion_matrix(Y_test, y_test_pred.T)
print("混淆矩阵为：",cm)
```

```python
TP = cm[0,0]
FN = cm[0,1]
FP = cm[1,0]
TN = cm[1,1]
ALL = sum(sum(cm))
```

```python
#输出评价指标
accuracy = (TP+TN)/ALL
print("Accuracy:", accuracy)
precision = TP/(TP+FP)
print("Precision:", precision)
recall = TP/(TP+FN)
print("Recall:", recall)
f1 = 2*TP/(2*TP+FP+FN)
print("F1:", f1)
```

```python
#绘制评价指标柱状图
data = [accuracy, precision, recall, f1]
labels = ["Accuracy", "Precision", "Recall", "F1"]
plt.bar(range(len(data)), data, tick_label=labels, color='g')
plt.savefig('./D4.svg')
plt.show()
```
